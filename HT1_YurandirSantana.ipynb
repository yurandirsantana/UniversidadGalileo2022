```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Funcion_Imagen(imagen, CodigoColor):

  img = cv2.imread(imagen)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Conversion de BGR a RGB

  # Dimensiones

  alto = img.shape[0]
  ancho = img.shape[1]

  lienzo = np.zeros((alto, ancho, 3))

  for i in range(0, alto):
    
    for j in range(0, ancho):
        pixel = img[i, j]
        
        blue = pixel[2]
        green = pixel[1]
        red = pixel[0]
        
        if(CodigoColor == 1):
          lienzo[i, j] = [blue, 0, 0]
        
        if(CodigoColor == 2):
          lienzo[i, j] = [0, green, 0]
        
        if(CodigoColor == 3):
          lienzo[i, j] = [0, 0, red]
        
        if(CodigoColor == 10):
          lienzo[i, j] = [0, green, red]
        
        if(CodigoColor == 20):
          lienzo[i, j] = [blue, green, 0]
        
        if(CodigoColor == 30):
          lienzo[i, j] = [blue, 0, red]

  cv2.imwrite("NuevaImagen.jpg", lienzo)
  NuevaImagen = cv2.imread("NuevaImagen.jpg")
  NuevaImagen = cv2.cvtColor(NuevaImagen, cv2.COLOR_BGR2RGB)
  return NuevaImagen
```

```python
Nueva_Imagen = "Francia.jpg"
plt.imshow(Funcion_Imagen(Nueva_Imagen,30))
```

```python
def ImagenEnColor(Imagen1, Imagen2, Imagen3):
    
    img1 = cv2.imread(Imagen1)
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) # Conversion de BGR a RGB

    img2 = cv2.imread(Imagen2)
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB) # Conversion de BGR a RGB

    img3 = cv2.imread(Imagen3)
    img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB) # Conversion de BGR a RGB
    
    # Dimensiones

    alto = img1.shape[0]
    ancho = img1.shape[1]
    
    lienzo = np.zeros((alto, ancho, 3))

    for i in range(0, alto):

        for j in range(0, ancho):

            pixelR = img1[i, j]
            pixelG = img2[i, j]
            pixelB = img3[i, j]

            RedR = pixelR[2] 
            GreenG = pixelG[1] 
            BlueB = pixelB[0]

            lienzo[i, j] = [BlueB, GreenG, RedR]
        
    
    cv2.imwrite("ImagenOG.jpg", lienzo)
    ImagenOG = cv2.imread("ImagenOG.jpg")
    ImagenOG = cv2.cvtColor(ImagenOG, cv2.COLOR_BGR2RGB)
    return ImagenOG        
```

```python
Nueva_ImagenR = "imagen1/imagen1_salida_gray_rojo.jpg"
Nueva_ImagenG = "imagen1/imagen1_salida_gray_verde.jpg"
Nueva_ImagenB = "imagen1/imagen1_salida_gray_azul.jpg"
plt.imshow(ImagenEnColor(Nueva_ImagenR,Nueva_ImagenG,Nueva_ImagenB))
```

```python
def EscalaGrisesPonderado(Imagen1):
    
    img = cv2.imread(Imagen1)
    alto = img.shape[0]
    ancho = img.shape[1]
    
    gris_ponderado = np.zeros((alto, ancho, 1))
    
    for i in range(0, alto):
        
        for j in range(0, ancho):

            pixel = img[i, j]

            blue = pixel[2]
            green = pixel[1]
            red = pixel[0]

            # Escala de grises ponderado

            gris_ponderado[i, j] = int(0.299*blue + 0.587*green + 0.114*red) 

    cv2.imwrite("gris_ponderado.jpg", gris_ponderado)
    monocromo = cv2.imread("gris_ponderado.jpg")
    monocromo = cv2.cvtColor(monocromo, cv2.COLOR_BGR2RGB)
    plt.imshow(monocromo)
    plt.show()
```

```python
Nueva_Imagen = "Francia.jpg"
```

```python
EscalaGrisesPonderado(Nueva_Imagen)
```

**Espacio de Color HSV**

El espacio de color HSV es una representaci√≥n tridimensional de color que se basa en los componentes de matiz, saturaci√≥n y valor.

Es altamente comparado con RGB, pero se diferencia en que es una representaci√≥n mas cercana y real a como lo seres humanos perciben los colores y las propiedades de cada uno de estos.

El **proceso de mapeo** es a trav√©s de los tres componentes que pueden variar en diferentes rangos:


*   **Matiz**: Este es representado como el valor de un √°ngulo que puede ir desde los cero a los 360 grados, y donde cada valor representa un calor.


*   **Saturaci√≥n**: Este es representado como la distancia que hay hasta el eje blanco-negro y los valores posibles van desde el cero al 100%.


*   **Valor**: Este es representado como la altura en el eje blanco-negro, y los valores posibles van desde el cero al 100%.
